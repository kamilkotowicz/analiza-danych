# Wczytywanie zbioru danych

```python
from sklearn import datasets
import numpy as np
import pandas as pd

iris = datasets.load_iris()
iris = pd.read_csv('Iris.csv', sep=',')
```

```python
df = pd.DataFrame(iris)
df
```

# Podział zbioru danych ze względu na gatunek kwiatu

```python
df0 = df[df['Species'] == 'Iris-versicolor'].drop(columns = ['Id', 'Species'])
df0
```

```python
df1 = df[df['Species'] == 'Iris-virginica'].drop(columns = ['Id', 'Species'])
df1
```

# Projekcja danych

```python
def project_data(df):
    covariance_matrix = df.cov()
    values, vectors = np.linalg.eig(covariance_matrix)
    #top_vectors = vectors[:2].transpose()
    top_vectors = vectors[:, :2]
    projected = df.dot(top_vectors)
    return projected
```

```python
projected0 = project_data(df0)
projected0
```

```python
projected1 = project_data(df1)
projected1
```

# Wizualizacja zrzutowanych macierzy kowariancji przed wykonaniem diagonalizacji

```python
import matplotlib.pyplot as plt

plt.scatter(projected0[0], projected0[1])
plt.scatter(projected1[0], projected1[1], c='red')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.grid()
plt.show()
```

# Wyznaczenie macierzy kowariancji dla zrzutowanych danych

```python
A = projected0.cov()
A
```

```python
B = projected1.cov()
B
```

```python
lambdas1, K1 = np.linalg.eig(A)
lambdas1, K1
```

```python
A1 = np.diag(lambdas1)
A1
```

```python
A1_powered = np.diag(lambdas1 ** (-1/2))
A1_powered
```

```python
whitening_matrix = np.dot(A1_powered, K1.transpose())
whitening_matrix
```

```python
y_values = np.dot(whitening_matrix, projected0.transpose())
y_values
```

```python
y_values2 = np.dot(whitening_matrix, projected1.transpose())
y_values2
```

```python
left = np.dot(whitening_matrix, B)
right = np.dot(K1, A1_powered)
E = np.dot(left, right)
E
```

```python
lambdas2, K2 = np.linalg.eig(E)
K2
```

```python
z_values = np.dot(K2.transpose(), y_values)
z_values
```

```python
z_values2 = np.dot(K2.transpose(), y_values2)
z_values2
```
