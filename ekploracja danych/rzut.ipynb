# Rzutowanie wektorów na hiperpłaszczyznę

```python
import numpy as np
import matplotlib.pyplot as plt

def project_vector_onto_hyperplane(vector, normal_vector, threshold):
    distance = np.dot(vector, normal_vector) - threshold
    norm_square = np.dot(normal_vector, normal_vector)
    projection = vector - distance / norm_square * normal_vector
    return projection
```

Przykładowe dane 2D

```python
vector_to_project = np.array([1, 3])
hyperplane_normal = np.array([2, 1])
threshold = 1
```

Rzutowanie wektora 2D na hiperpłaszczyznę

```python
projected_vector = project_vector_onto_hyperplane(vector_to_project, hyperplane_normal, threshold)
print(projected_vector)
```

Wykres dla wektorów 2D

```python
a, b, c = hyperplane_normal[0], hyperplane_normal[1], threshold

if len(vector_to_project) == 2:
    plt.figure(figsize=(7, 7))

    # Wykres linii siatki
    plt.grid(True, linestyle='--', alpha=0.7)

    # Wykres hiperpłaszczyzny
    x = np.linspace(-5, 5, 100)
    y = (-hyperplane_normal[0] * x + threshold) / hyperplane_normal[1]
    plt.plot(x, y, 'g--', label=f"{hyperplane_normal[0]}x + {hyperplane_normal[1]}y  = {threshold}")

    # Wykres wektora oryginalnego
    plt.quiver(0, 0, vector_to_project[0], vector_to_project[1], angles='xy', scale_units='xy', scale=1, color='b', label="Wektor oryginalny")

    # Wykres wektora normalnego
    plt.quiver(0, 0, hyperplane_normal[0], hyperplane_normal[1], angles='xy', scale_units='xy', scale=1, color='r', label="Wektor normalny")

    # Wykres punktu po rzutowaniu
    plt.scatter(projected_vector[0], projected_vector[1], color='r', marker='x', s=100, label="Punkt po rzutowaniu")

    plt.xlim(-5, 5)
    plt.ylim(-5, 5)
    plt.title("Rzutowanie wektora na hiperpłaszczyznę")
    plt.legend()

    plt.show()
```

# Odbicie wektora od hiperpłaszczyzny

```python
def reflect_vector_onto_hyperplane(vector, normal_vector, threshold):
    projected_vector = project_vector_onto_hyperplane(vector_to_reflect, hyperplane_normal, threshold)
    difference = (projected_vector - vector) * 2
    reflection = vector + difference
    return reflection
```

Przykładowe dane 2D

```python
vector_to_reflect = np.array([1, 3])
hyperplane_normal = np.array([2, 1])
threshold = 1
```
