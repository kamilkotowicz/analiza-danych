# Wczytywanie zbioru danych

```python
from sklearn import datasets
import numpy as np
import pandas as pd

iris = datasets.load_iris()
iris
```

```python
df = pd.DataFrame(iris.data)
df['target'] = iris.target
df
```

# Podział zbioru danych ze wzlędu na gatunek kwiatu

```python
df1 = df[df['target'] == 1].drop(columns = ['target'])
df1
```

```python
df2 = df[df['target'] == 2].drop(columns = ['target'])
df2
```

# Projekcja danych

```python
def project_data(df):
    covariance_matrix = df.cov()
    values, vectors = np.linalg.eig(covariance_matrix)
    top_vectors = vectors[:, :2]
    projected = df.dot(top_vectors)
    return projected
```

```python
projected1 = project_data(df1)
projected1
```

```python
projected2 = project_data(df2)
projected2
```

# Wizualizacja danych zrzutowanych na przestrzeń dwuwymiarową

```python
import matplotlib.pyplot as plt

plt.scatter(projected1[0], projected1[1])
plt.scatter(projected2[0], projected2[1], c='red')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.grid()
plt.show()
```

# Wizualizacja danych nachodzących na siebie

```python
projected3 = projected2.copy()
projected3[0] = projected2[0]-1
projected3[1] = projected2[1]+0.75
plt.scatter(projected1[0], projected1[1])
plt.scatter(projected3[0], projected3[1], c='red')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.grid()
plt.show()
```

# Obliczenie fisherowskiego kierunku

```python
mean1 = np.mean(projected1)
mean1
```

```python
mean2 = np.mean(projected2)
mean2
```

```python
mean3 = np.mean(projected3)
mean3
```

```python
def get_scatter_matrix(data, mean):
    scatter_matrix =  np.array([[0.0,0.0],[0.0,0.0]])
    for index, row in data.iterrows():
        term = np.outer(row-mean, row-mean)
        scatter_matrix += term
    return scatter_matrix
```

```python
scatter1 = get_scatter_matrix(projected1, mean1)
scatter1
```

```python
scatter2 = get_scatter_matrix(projected2, mean2)
scatter2
```

```python
scatter3 = get_scatter_matrix(projected3, mean3)
scatter3
```

```python
Sw = scatter1 + scatter2
Sw
```

```python
Sw2 = scatter1 + scatter3
Sw2
```

```python
from numpy.linalg import inv
Sw_inw = inv(Sw)
Sw_inw
```

```python
Sw_inw2 = inv(Sw2)
Sw_inw2
```

```python
w = np.dot(Sw_inw, (mean2 - mean1).transpose())
w
```

```python
w2 = np.dot(Sw_inw2, (mean3 - mean1).transpose())
w2
```

# Rzutowanie danych na wektor fisherowski

```python
def project_point_onto_vector(point, vector):
    norm = np.linalg.norm(vector)
    vector /= norm
    projection = np.dot(point, vector) * vector
    return projection
```

```python
def get_projected_points(data, w):
    double_projected = []
    for index, row in data.iterrows():
        projected_point = project_point_onto_vector(row, w)
        double_projected.append(projected_point)
    double_projected = np.array(double_projected)
    return double_projected
```

```python
double_projected1 = get_projected_points(projected1, w)
double_projected1 = double_projected1[double_projected1[:, 0].argsort()]
double_projected1
```

```python
double_projected2 = get_projected_points(projected2, w)
double_projected2 = double_projected2[double_projected2[:, 0].argsort()]
double_projected2
```

```python
double_projected3 = get_projected_points(projected3, w2)
double_projected3 = double_projected3[double_projected3[:, 0].argsort()]
double_projected3
```

```python
double_projected4 = get_projected_points(projected1, w2)
double_projected4 = double_projected4[double_projected4[:, 0].argsort()]
double_projected4
```

# Znalezienie punktu podziału dla nierozdzielnych danych

```python
arr1_num = np.ones_like(double_projected4[:, 0])
arr2_num = 2 * np.ones_like(double_projected3[:, 0])
combined = np.column_stack((np.concatenate((double_projected4[:, :2], double_projected3[:, :2])), np.concatenate((arr1_num, arr2_num))))
combined = combined[combined[:, 0].argsort()]
combined
```

```python
jedynki = 0
pref = []
for i in range(len(combined)):
    if combined[i][2] == 1:
        jedynki += 1
    pref.append(jedynki)

best_mistakes = len(combined)
best_index = i
for i in range(len(combined)):
    mistakes = (i+1 - pref[i]) + (pref[-1] - pref[i])
    if mistakes < best_mistakes:
        best_mistakes = mistakes
        best_index = i
best_index, best_mistakes, combined[best_index]
split = (combined[best_index] + combined[best_index+1])[0:2]/2
split
```

```python
plt.scatter(double_projected4[:, 0], double_projected4[:, 1])
plt.scatter(double_projected3[:, 0], double_projected3[:, 1], c='red')
plt.scatter(split[0], split[1], c='black')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.grid()
plt.show()
```
